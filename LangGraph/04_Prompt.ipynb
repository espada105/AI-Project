```python
template = "{country}의 수도는 어디인가요?"

prompt = PromptTemplate(
    template = template,
    input_variables = ["country"],

)

prompt
```

```python
prompt.format(country = "Korea")
```

```python
template = "{country1} 와 {country2} 의 수도는 어디인가요?"

prompt = PromptTemplate(
    template = template,
    input_variables = [ "country1"],
    partial_variables = {
        "country2": "america"
    }
)

prompt
```

```python
prompt.format(country1 = "korea")
```

```python
prompt_partial = prompt.partial(country2 = "canada")

prompt_partial
```

```python
prompt_partial.format(country1 = "korea")

chain = prompt_partial | llm

chain.invoke("korea").content
```

```python
chain.invoke({"country1" : "korea", "country2" : "australia"}).content
```

```python
from datetime import datetime

datetime.now().strftime("%B %d")
```

```python
def get_today():
    return datetime.now().strftime("%B %d")
```

```python
prompt = PromptTemplate(
    template = "오늘의 날짜는 {today} 입니다. 오늘이 생일인 유명인 {n}명을 나열해 주세요. 생년월일을 표기해 주세요",
    input_variables = ["n"],
    partial_variables = {
        "today": get_today
    }, 
)
```

```python
prompt.format(n = 3)
```

```python
chain = prompt | llm

print(chain.invoke(3).content)
```

```python
print(chain.invoke({"today": "Jan 02", "n": 3}).content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant, your name is {name}"),
        ("human", "hello, how are you?"),
        ("ai", "hello what can"),
        ("human", "{user_input}"),
    ]
)

messages = chat_template.format_messages(
    name = "teddy",
    user_input = "what is my name?"
)

messages

```
