```python
# stream: 응답의 청크를 스트리밍
# invoke: 입력에 대해 체인을 호출
# batch: 입력목록에 대해 체인을 호출

# astream: 비동기적으로 응답의 청크를 스트리밍
# ainvoke: 비동기적으로 입력에 대해 체인을 호출
# abatch: 비동기적으로 입력 목록에 대해 체인을 호출
# astream_log: 최종 응답뿐만 아니라 발생하는 중간 단게를 스트리밍한다.
```

```python
from dotenv import load_dotenv

load_dotenv
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

model = ChatOpenAI(model="gpt-3.5-turbo")

prompt = PromptTemplate.from_template("{topic}에 대해서 알려주세요")

chain = prompt | model | StrOutputParser()
```

```python
for token in chain.stream({"topic": "multimodal"}):
    print(token, end="", flush = True)
```

```python
for token in chain.stream({"topic": "multimodal"}):
    print(token, end="", flush = False)
```

```python
chain.invoke({"topic": "multimodal"})
```

```python
chain.batch([{"topic":"multimodal"}, {"topic":"bitcoin history"}])
```

```python
chain.batch(
    [
        {"topic":"multimodal"},
        {"topic":"bitcoin history"},
        {"topic":"airconditional history"},
        {"topic":"AI"},
        {"topic":"AI agent"},
    ],
    config = {"max_concurrency": 3},
)
```

```python
async for token in chain.astream({"topic": "multimodal"}):
    print(token, end = "", flush = True)
```

```python
my_process = chain.ainvoke({"topic": "langchain difference between invoke and stream"})

await my_process
```

```python
my_abatch = chain.abatch(
    [{"topic": "multimodal"}, 
    {"topic": "bitcoin history"}, 
    {"topic": "airconditional history"}]
)
```

```python
await my_abatch
```

```python
from langchain_core.runnables import RunnableParallel

chain1 = (
    PromptTemplate.from_template("{country}의 수도는 어디야?")
    | model
    | StrOutputParser()    
)

chain2 = (
    PromptTemplate.from_template("{country}의 면적은 얼마야?")
    |model
    |StrOutputParser()
)

combined = RunnableParallel(capital = chain1, area = chain2)
```

```python
chain1.invoke({"country": "korea"})
chain2.invoke({"country": "america"})

combined.invoke({"country": "korea"})
```

```python
chain1.invoke({"country": "korea"})
```

```python
chain2.invoke({"country": "america"})
```

```python
combined.invoke({"country": "korea"})
```

```python
chain1.batch([{"country": "korea"}, {"country": "america"}])
```

```python
chain2.batch([{"country": "korea"}, {"country": "america"}])
```
