```python
from langchain_teddynote.messages import stream_response
from langchain_core.prompts import PromptTemplate
```

```python
template = "{country}의 수도는 어디임?"
prompt_template = PromptTemplate.from_template(template)
prompt_template
```

```python
prompt = prompt_template.format(country = "korea")
prompt
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model = "gpt-3.5-turbo",
    max_tokens = 2048,
    temperature = 0.1,
)
```

```python
prompt = prompt_template.from_template(f"{topic}에 대해서 알려주세요")

model = ChatOpenAI()

chain = prompt | model 
```

```python
input = {"topic":"에어컨 역사"}
```

```python
chain.invoke(input)
```

```python
answer = chain.stream(input)

stream_response(answer)
```

```python
from langchain_core.output_parsers import StrOutputParser

templete = """
    you are a helpful assistant.
    you are a expert of history.
    you are given a question and you need to answer the question.

    format:
    1. only answer the question for elementary school student in english.
    2. answer in korean.

"""

prompt_template = PromptTemplate.from_template(templete)

output_parser = StrOutputParser()

chain = prompt | model | output_parser

answer = chain.stream({"question": "airconditional history"})

stream_response(answer)
```
